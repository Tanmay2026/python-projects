```python
# import libraries
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
import joblib
```

## Loading and Processing Data

```python
data = pd.DataFrame(load_iris().data)
data['4'] = load_iris().target
columns = ['Sepal Length(cm)','Sepal width(cm)','Petal Length(cm)','Petal Width(cm)','types']
data.columns = columns
data.head()
```

```python
data.info()
```

```python
data.describe()
```

## Check for null values

```python
data.isnull().any()
```

## Visualize Dataset

```python
sns.pairplot(data, hue='types', height=2.5);
```

## Building Model

### Processing Data

```python
features, label = data[['Sepal Length(cm)','Sepal width(cm)','Petal Length(cm)','Petal Width(cm)']], data['types']
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size = 0.1)
```

```python
"Training Data Size : {}".format(x_train.shape[0])
```

```python
"Test Data Size : {}".format(x_test.shape[0])
```

### Training Logistic Model

```python
model = LogisticRegression(max_iter=300)
model.fit(x_train, y_train)
```

## Calculate Model Accuracy

```python
model_score = model.score(x_test, y_test)
```

```python
"Logistic Model Score for IRIS Flower Classification : {}".format(model_score)
```

```python
df = pd.DataFrame()
df['Actual'] = y_test
df['Predictions'] = model.predict(x_test)
df
```

## Saving the model

```python
modelName = "iris_flower_classifier_logistic_regression.sav"
joblib.dump(model, modelName)
```
